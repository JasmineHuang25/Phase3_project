# EDA and more data preparation

Please note the main 'df.csv' is not the original dataset. The 3 original datasets are owned and published by Chicago Police Department.

- Traffic Crashes - Crashes: 681,683 rows and 49 columns
- Traffic Crashes - Vehicles: 1,494,816 rows and 30 columns
- Traffic Crashes - People: 1,389,220 rows and 72 columns

The 3 original datasets have 151 columns/features and 3,565,719 rows/observations.  After first initial data cleaning and selecting, the merged 'df.csv' has 14 columns and 339,815 rows. Its memory usage is 36.3MB.  The 14 columns are:
- 'CRASH_DATE'
- 'DEVICE_CONDITION'
- 'POSTED_SPEED_LIMIT'
- 'WEATHER_CONDITION'
- 'LIGHTING_CONDITION'
- 'ROADWAY_SURFACE_COND'
- 'ROAD_DEFECT'
- 'DAMAGE'
- 'ZIPCODE'
- 'SEX'
- 'AGE'
- 'MAKE'
- 'MODEL'
- 'VEHICLE_YEAR'

During the first EDA, I have inspected and cleaned some columns.  Here I will be doing the same for the rest of columns, 'CRASH_DATE', 'DAMAGE', 'ZIPCODE', 'SEX', 'AGE', 'MAKE', 'MODEL', 'VEHICLE_YEAR'.

#### 3. ZIPCODE

Trim the long tails and outliers: drop 'ZIPCODE' if its value_counts is less than 600.

```python
zipcode_count = df['ZIPCODE'].value_counts()
zipcode_count
```

```python
zipcod_dict = zipcode_count.to_dict()

zipcode_drop_list= []

for k in zipcod_dict:
    if zipcod_dict[k] < 600:
        zipcode_drop_list.append(k)

df = df[df.ZIPCODE.isin(zipcode_drop_list) == False]
```

```python
len(df['ZIPCODE'].value_counts())
```

```python
fig, ax = plt.subplots(figsize=[20,10])
df['ZIPCODE'].value_counts().plot(kind='bar')
plt.xticks(rotation=45, ha='right')
plt.title("ZIPCODE")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=[20,10])
sns.barplot(x=df['ZIPCODE'].value_counts().index, y=df['ZIPCODE'].value_counts(), color='purple',alpha=0.8)
plt.xticks(rotation=45, ha='right')
plt.show();
```

#### 7. MODEL
Trim the long tails and outliers: drop 'MODEL' if its value_counts is less than 100.
